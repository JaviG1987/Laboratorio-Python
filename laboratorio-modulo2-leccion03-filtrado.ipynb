# Laboratorio Filtrado de Datos

En este laboratorio usaremos el DataFrame creado en la lección anterior resultado de la unión de los conjuntos de datos compartidos contigo.

```python
import pandas as pd
pd.set_option ("display.max_columns", None)
```

```python
df_mergeado_netflix = pd.read_csv("data_mergeado.csv", index_col=0)
df_mergeado_netflix.head()
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
filtro_2021 = df_mergeado_netflix["release_year"] == 2021
fitro_type = df_mergeado_netflix ["type"] == "Movie"
df_2021_movie = df_mergeado_netflix.loc [filtro_2021 & fitro_type, ["Title","type", "release_year"]]
df_2021_movie
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
filtro_duracion = df_mergeado_netflix ["duration"] == "90 min"
df_duracion90 = df_mergeado_netflix.loc [filtro_duracion,["Title", "duration"]]
df_duracion90
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
filtro_pais = df_mergeado_netflix ["country"] == "India"
filtro_release_year = df_mergeado_netflix ["release_year"] > 2018
df_India_2018 = df_mergeado_netflix.loc [filtro_pais & filtro_release_year, ["Title", "release_year", "country"]]
df_India_2018
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
filtro_rating = df_mergeado_netflix ["rating"] == "TV-14"
filtro_rating2 = df_mergeado_netflix ["rating"] == "R"
df_rating = df_mergeado_netflix.loc [filtro_rating | filtro_rating2,["Title","rating"]]
df_rating
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
filtro_director = df_mergeado_netflix ["director"] == "Christopher Nolan"
cuantos_Nolan = filtro_director.sum()
print(f" El directo Christopher Nolan ha dirigido {cuantos_Nolan} titulo de los que están en la lista")
```

```python
filtro_director2 = df_mergeado_netflix ["director"] == "Martin Scorsese"
cuantos_Martin = filtro_director2.sum()
print(f" El directo Martin Scorsese ha dirigido {cuantos_Martin} titulo de los que están en la lista")
```

```python
df_director = df_mergeado_netflix.loc[filtro_director | filtro_director2, ["Title","director"]]
df_director
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
filtro_pais2 = df_mergeado_netflix ["country"] == "United States"
filtro_pais3 = df_mergeado_netflix ["country"] == "United Kingdom"
filtro_pais4 = df_mergeado_netflix ["country"] == "Canada"
filtro_mas_language = df_mergeado_netflix["Language"].str.contains("/", na=False)
df_maslenguaje = df_mergeado_netflix.loc[filtro_pais2 | filtro_pais3 | filtro_pais4, ["Title", "country", "Language"]]
df_maslenguaje_2 = df_maslenguaje.loc [filtro_mas_language,["Title","Language"]]
df_maslenguaje_2
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
filtro_2015 = df_mergeado_netflix["release_year"] >= 2015
filtro_2020 = df_mergeado_netflix["release_year"] <= 2020
df_discos_lanzados = df_mergeado_netflix.loc[filtro_2015 & filtro_2020, ["Title","release_year"]]
df_discos_lanzados
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
filtro_IMDb = df_mergeado_netflix ["IMDB Score"]> 6.5
filtro_IMDb2 = df_mergeado_netflix ["IMDB Score"] < 8.0
filtro_duracion2 = df_mergeado_netflix ["duration"] == "NaN"
df_puntuacion_duracion = df_mergeado_netflix.loc [filtro_IMDb & filtro_IMDb2 & filtro_duracion2,["Title","duration","IMDB Score"]]
df_puntuacion_duracion
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
filtro_durancion90 = df_mergeado_netflix ["duration"] == "90 min"
filtro_durancion150 = df_mergeado_netflix ["duration"] == "150 min"
df_duracion_90_150 = df_mergeado_netflix.loc [filtro_durancion90 | filtro_durancion150,["Title","duration"]]
df_duracion_90_150
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
filtro_love = df_mergeado_netflix["Title"].str.contains("Love")
filtro_type = df_mergeado_netflix["type"] == "TV Show"
df_Love_TVShow = df_mergeado_netflix.loc [filtro_love & filtro_type, ["Title", "type"]]
df_Love_TVShow
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
filtro_apellido = df_mergeado_netflix ["director"].str.contains("Johnson", na=False)
df_apellido_director = df_mergeado_netflix.loc [filtro_apellido,["Title", "director"]]
df_apellido_director
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
filtro_documental = df_mergeado_netflix ["listed_in"] == "Documentaries"
filtro_actor = df_mergeado_netflix ["cast"].str.contains ("Mike Tyson",na=False)
df_documental_Tyson = df_mergeado_netflix.loc[filtro_documental & filtro_actor, ["Title","cast","listed_in"]]
df_documental_Tyson
```

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
filtro_title = df_mergeado_netflix["Title"].notna()
filtro_country = df_mergeado_netflix["country"].notna()
filtro_IMDB_Score = df_mergeado_netflix["IMDB Score"].notna()
df_informacion_completa = df_mergeado_netflix.loc[filtro_title & filtro_country & filtro_IMDB_Score, ["Title", "country", "IMDB Score"]]
df_informacion_completa
```
