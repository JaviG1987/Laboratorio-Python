## Ejercicio 1. Definición de la clase Cancion
1.1. Define  la clase Cancion, la cual va a tener 4 atributos básicos. (título,  artista, album, genero y duración)

1.2. Crea un método para la clase Cancion que nos muestre la información de la canción.

1.3. Crea un método que nos permita comparar la duración de dos canciones y nos indique cuál de las dos es más larga.

1.4. Crea un método que nos permita calcular la popularidad de una canción en función de varios criterios.

- La puntuación base debe ser de 50.
    
- Las canciones de entre 2 y 4 minutos de duración son más populares, por lo que suman 20 puntos. Las demás restan 10.  

- Las canciones de los géneros pop, rock, hip-hop y electrónicas son populares, por lo que suman 20 puntos. El resto de géneros restan 5 puntos.

- Si una canción pertenece a un álbum suma 10 puntos.

1.5. Crea dos instancias de la clase canción con la siguiente info: ("Stairway to Heaven", "Led Zeppelin", "Led Zeppelin IV", 8.02, "Rock") y ("What's Going On", "Marvin Gaye", "What's Going On", 3.53, "Soul")

1.6. Comprueba el funcionamiento de cada uno de los métodos de la clase



````python
class cancion:
    def __init__ (self, titulo, artista, album, duracion, genero):

        self.titulo = titulo
        self.artista = artista
        self.album = album
        self.genero = genero
        self.duracion = duracion

    def info_cancion (self):
        return (f"EL titulo de la cancion es {self.titulo}\n"
                f"EL artista de la cancion es {self.artista}\n"
                f"EL album de la cancion es {self.album}\n"
                f"EL genero de la cancion es {self.genero}\n"
                f"La duracion de la cancion es {self.duracion}")

    def comparacion_canciones (self, cancion_comparar):
        if self.duracion > cancion_comparar.duracion:
            return (f"La canción {self.titulo} tiene más duración que la canción {cancion_comparar.titulo}")
        else:
            return (f"La canción {cancion_comparar.titulo} tiene más duración que la canción {self.titulo}")
        

    def puntuacion_cancion (self):
        puntuacion = 50
        if self.duracion >= 2 and self.duracion <= 4:
            puntuacion += 20
        else:
            puntuacion -=10
        if self.genero == "Pop" or self.genero == "Rock" or self.genero == "Hip-Hop" or self.genero == "Electronica":
            puntuacion += 20
        else:
            puntuacion -=10
        if self.album != " ":
            puntuacion += 10
        else:
            puntuacion = puntuacion
        return (f"la puntuacion de la cancion es {puntuacion}\n")

cancion1 = cancion("Stairway to Heaven", "Led Zeppelin", "Led Zeppelin IV", 8.02, "Rock")
cancion2 = cancion("What's Going On", "Marvin Gaye", "What's Going On", 3.53, "Soul")

print (f"{cancion1.info_cancion()}")
print (f"{cancion1.puntuacion_cancion()}")

print (f"{cancion2.info_cancion()}")
print (f"{cancion2.puntuacion_cancion()}")

print (f"{cancion1.comparacion_canciones(cancion2)}")
````

````python
class colaboracion (cancion):
    def __init__ (self, titulo, artista, album, duracion, genero, artistas_colaboradores):
        super ().__init__(titulo, artista, album, duracion, genero)

        self.artistas_colaboradores = artistas_colaboradores

    def info_cancion (self):
            return (f"EL titulo de la cancion es {self.titulo}\n"
                    f"EL artista de la cancion es {self.artista}\n"
                    f"EL album de la cancion es {self.album}\n"
                    f"EL genero de la cancion es {self.genero}\n"
                    f"La duracion de la cancion es {self.duracion}\n"
                    f"Los artistas colaboradores son {self.artistas_colaboradores}")
        
    def colaboracion_nocolaboracion (self):
            if len (self.artistas_colaboradores) != 0:
                return (f"la cancion {self.titulo} es una colaboración")
            else:
                return (f"la cancion {self.titulo} no es una colaboración")
        
    def numero_colaboradores (self):
            contador = []
            cuantos_colaboradores = 0
            for i in self.artistas_colaboradores:
                contador.append (i)
                cuantos_colaboradores = len(contador)
            return (f"La cancion tiene {cuantos_colaboradores} colaboradores")
        
    def puntuacion (self):
            puntuacion = 50
            if self.duracion >= 2 and self.duracion <= 4:
                puntuacion += 20
            else:
                puntuacion -=10
            if self.genero == "Pop" or self.genero == "Rock" or self.genero == "Hip-Hop" or self.genero == "Electronica":
                puntuacion += 20
            else:
                puntuacion -=10
            if self.album != " ":
                puntuacion += 10
            else:
                puntuacion = puntuacion
            if self.artistas_colaboradores != " ":
                puntuacion += 5
            return (f"la puntuacion de la cancion es {puntuacion}\n")
        
    def colaborador_principal (self):
            if len (self.artistas_colaboradores) != 0:
                 return (f"El colaborador principal es {self.artistas_colaboradores[0]}")

cancion3 = colaboracion("Meant to Be", "Bebe Rexha", "Expectations", 2.44, "Country Pop", ["Florida Georgia Line"])
````
